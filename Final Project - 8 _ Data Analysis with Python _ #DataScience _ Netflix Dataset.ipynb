Q. 6. Show Top 10 Directors, who gave the highest number of TV Shows & movies to Netlix? 

```python
# value_counts()
```

```python
# Group the data by 'Director' and 'Category', then count the number of occurrences
director_counts = data.groupby(['Director', 'Category']).size().unstack().fillna(0)

# Calculate the total count of TV shows and movies for each director
director_counts['Total'] = director_counts['Movie'] + director_counts['TV Show']

# Sort the directors by the total count in descending order
top_directors = director_counts.sort_values(by='Total', ascending=False)

# Get the top 10 directors
top_10_directors = top_directors.head(10)                                                             
```

```python
top_10_directors
```

Q. 8 in How many movies /shows, Tom Cruise was cast?

```python
# Filter the DataFrame to include only records where Tom Cruise is in the 'Cast' column
tom_cruise_movies_shows = data[data['Cast'].str.contains('Tom Cruise', case=False, na=False)]

# Calculate the total number of movies or shows in which Tom Cruise was cast
count_tom_cruise_movies_shows = len(tom_cruise_movies_shows)

("Tom Cruise was cast in", count_tom_cruise_movies_shows, "movies/shows.")
```

Q. 9 what are the different Ratings defined by Netflix

Q. 9.2 How many TV Show  got the 'R' rating , after year 2018?

```python
# Convert the 'Release_Date' column to a datetime object
data['Release_Date'] = pd.to_datetime(data['Release_Date'])

# Filter the DataFrame to include only TV shows with 'R' rating released after 2018
r_rated_tv_shows_after_2018 = data[(data['Category'] == 'TV Show') & (data['Rating'] == 'R') & (data['Release_Date'].dt.year > 2018)]

# Get the count of 'R' rated TV shows released after 2018
count_r_rated_tv_shows_after_2018 = len(r_rated_tv_shows_after_2018)

("Number of 'R' rated TV shows released after 2018:", count_r_rated_tv_shows_after_2018)
```

Q.10 What is the Maximum duration of a Movie/Show on Netflix? 

```python
# Convert the 'Duration' column to a string and remove any non-numeric characters
data['Duration'] = data['Duration'].str.replace('\D', '', regex=True)

# Convert the 'Duration' column to numeric
data['Duration'] = pd.to_numeric(data['Duration'], errors='coerce')

# Find the maximum duration
max_duration = data['Duration'].max()

# Display the maximum duration
print("Maximum Duration of a Movie/Show on Netflix:", max_duration, "minutes")

```

```python
# Q.11 Wich individual country has the highest No of TV shows? 
```

```python
# Filter the DataFrame to include only TV shows
tv_shows = data[data['Category'] == 'TV Show']

# Count the number of TV shows by country
tv_shows_by_country = tv_shows['Country'].value_counts()

# Get the country with the highest number of TV shows
country_with_highest_tv_shows = tv_shows_by_country.idxmax()
highest_tv_show_count = tv_shows_by_country.max()


(f"The country with the highest number of TV shows is {country_with_highest_tv_shows} with {highest_tv_show_count} TV shows.")
```
